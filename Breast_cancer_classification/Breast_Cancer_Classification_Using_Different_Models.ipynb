### Dataset 2

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df2.head()
```

```python
df2.columns
```

```python
df2_clean = df2.drop(['id', 'Unnamed: 32'], axis=1)
df2_clean['diagnosis'] = (df2_clean['diagnosis'] == 'M').astype(int)

X2 = df2_clean.drop('diagnosis', axis=1)
y2 = df2_clean['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2, y2, test_size=0.2, random_state=42)

scaler = StandardScaler()
X2_train_scaled = scaler.fit_transform(X2_train)
X2_test_scaled = scaler.transform(X2_test)
```

```python
# Knn
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X2_train_scaled, y2_train)
print(f'Accuracy Score: {knn_clf.score(X2_test_scaled, y2_test):.4f}')
```

```python
#logistic regression
lr_clf = LogisticRegression(max_iter=1000, random_state=42)
lr_clf.fit(X2_train_scaled, y2_train)
print(f'Accuracy Score: {lr_clf.score(X2_test_scaled, y2_test):.4f}')
```

```python
# Decision Tree Classifier
dt_clf = DecisionTreeClassifier(random_state=42, max_depth=10)
dt_clf.fit(X2_train_scaled, y2_train)
print(f'Accuracy Score: {dt_clf.score(X2_test_scaled, y2_test):.4f}')
```

```python
# Random Forest Classifier
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42, max_depth=10)
rf_clf.fit(X2_train_scaled, y2_train)
print(f'Accuracy Score: {rf_clf.score(X2_test_scaled, y2_test):.4f}')
```

```python
# SVM Classifier
svm_clf = SVC(kernel='rbf', random_state=42)
svm_clf.fit(X2_train_scaled, y2_train)
print(f'Accuracy Score: {svm_clf.score(X2_test_scaled, y2_test):.4f}')
```
